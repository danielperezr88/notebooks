```python
import numpy as np
# from sklearn import preprocessing, cross_validation, neighbors
import pandas as pd

df = pd.read_csv('churn-data.csv')

# PREPROCESAMOS LOS DATOS dejandolo todo en BINARIO 1, 0
# CUIDADO QUE CUANDO PANDA VE UN TRUE lo interpreta internamente como 1, idem FALSE  Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡

#df.replace('True',1,inplace=True)
df.replace('yes',1,inplace=True)
#df.replace('False',0,inplace=True)
df.replace('no',0,inplace=True)

df.head()
```

```python
# CUIDADO QUE CUANDO PANDA VE UN TRUE lo interpreta internamente como 1, idem FALSE  Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡Â¡
# LO COMPROBAMOS
df[df.Churn == 1]
```

```python
# otras funciones...
print(df.columns)
print(df.values)
df['Churn']
df.head()
```

```python
# creamos un df1 con los datos REALMENTE UTILES
df1 = df
del df1['State']
del df1['Area_Code']
del df1['Phone']
df1.head()
```

```python
# CLASIFICADOR montamos los arrays de NUMPY eliminando las columnas que no aportan correlacion
X = np.array(df1.drop(['Churn'], 1))
y = np.array(df1['Churn'])
```

```python
from sklearn import preprocessing, cross_validation, neighbors, svm
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2)

# utilizamos el support vector classifier  
clf = svm.SVC()

clf.fit(X_train, y_train)
confidence = clf.score(X_test, y_test)
print(confidence)
```

```python
# vamos a hacer predicciones :


# CAMBIAR LOS DATOS
#example_measures = np.array([[65,0,0,0,129.1,137,21.95,228.5,83,19.42,208.8,111,9.40,12.7,6,3.43,4]])


example_measures = np.array([X[10]])

#TRANSFORMAMOS esta lista mediante RESHAPE para que sea CONSUMIBLE por el clasificador clf
example_measures = example_measures.reshape(len(example_measures), -1)

prediction = clf.predict(example_measures)
print prediction
```

```python
# funcion GENERICA QUE DIBUJA UNA COMBINACION particular de las variable

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import style

def dibuja_grafica(i,j,k):

    style.use('ggplot')

    fig = plt.figure()
    ax1 = fig.add_subplot(111, projection='3d')


    x =  np.array(df_abandono.iloc[:,i])
    y =  np.array(df_abandono.iloc[:,j])
    z =  np.array(df_abandono.iloc[:,k])

    x2 = np.array(df_NO_abandono.iloc[:,i])
    y2 = np.array(df_NO_abandono.iloc[:,j])
    z2 = np.array(df_NO_abandono.iloc[:,k])

    ax1.scatter(x, y, z, c='g', label='ABANDONO', marker='o')
    ax1.scatter(x2, y2, z2, c ='r', label='LEALES', marker='o')
    
 
    ax1.set_xlabel(df_abandono.columns[i])
    ax1.set_ylabel(df_abandono.columns[j])
    ax1.set_zlabel(df_abandono.columns[k])
    
    #plt.title('AnÃ¡lisis de Tasa de Abandono Churn Rate Analysis Human Forecast 2016')
    #plt.legend()

    plt.show()  
    #plt.savefig('graf_'+df_abandono.columns[i]+df_abandono.columns[j]+df_abandono.columns[k]+'.png')

dibuja_grafica(1,2,4)
```
